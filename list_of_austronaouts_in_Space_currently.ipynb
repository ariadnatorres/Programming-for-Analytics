Import Libraries

```python
import requests
import json
import pandas as pd
```

Specify the URL of the API services

```python
response = requests.get('http://api.open-notify.org/astros.json') #this can be found on the top of the webpage, link on their website
```

```python
#check if request was sucessfull

if response.status_code == 200:

  #retrieve the json from response
  data = response.json()


  #format the json returned by the API
  #This will help us take the raw data and organize it so it is easier to read
  #formatted_json = json.dumps(data, sort_keys = True, indent = 10)

  #print(formatted_json)

  #Build a dataframe with contain the spacecrafts and the austronauts aboard such
  #everytime you are building a dataframe, you would want to have a list for each colum

  spacecraft_name = []
  astronaut_name = []

  #print all the names of all of the austronauts in space
  #we can notice that this list has dictionaries for each austronaut

  #lets try to retrieve ONLY the name of each austronaut
  for astronauts in data['people']:
    spacecraft_name.append(astronauts['craft'])
    astronaut_name.append(astronauts['name'])

  #Creating a series
  space_data = {'Spacecraft Name': spacecraft_name, 'Astronaut': astronaut_name}

  #Build a dataframe from the data collected
  df = pd.DataFrame(space_data)


```
