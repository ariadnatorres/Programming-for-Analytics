Import the libraries

```python
# Import requests, pandas and BeautifulSoup
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Define a function that can take a URL parameter, check the status code, get the content of the response, and create a dataframe

```python


# Define a function that takes a URL and a file name as parameters
def scrape_table(url):
  # Make a GET request to the URL
  response = requests.get(url)

  # Check the status code
  if response.status_code == 200:
    print(f"Status code: {response.status_code}")

    # Get the content of the response
    content = response.content

    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(content, "html.parser")

    # Find the table element by its id or class attribute
    # We need to inspect the web page to find the right attribute which in this case are table and example2
    table = soup.find("table", id="example2")

    # Read the table using pandas
    df = pd.read_html(str(table))[0]

    # Print the dataframe
    print(df)



    # Save the dataframe as a CSV file
    df.to_csv("population2.csv", index=False)

    # Return the dataframe
    return df

```

Scrape the contents using the function and saving it on df2 variable

```python
# scrape the contents using the function ans save it on df2
df2 = scrape_table("https://www.worldometers.info/world-population/population-by-country/")
```
